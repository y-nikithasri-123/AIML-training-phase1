LINE CHART

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=[20,25,35,55]
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Thur','Fri','Sat','Sun']
y=[170,120,250,190]
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()              
```
